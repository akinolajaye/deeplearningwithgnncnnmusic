```python
import numpy as np
import pandas as pd
```

```python
spotify_df = pd.read_csv("spotify_dataset.csv", on_bad_lines="skip")
```

```python
spotify_df
```

```python
spotify_df.columns
```

```python
spotify_df.drop(columns=[spotify_df.columns[-1]],inplace=True)
```

```python
spotify_df
```

```python
spotify_df.columns = spotify_df.columns.str.replace('"', '')
spotify_df.columns = spotify_df.columns.str.replace(' ', '')
```

```python
spotify_df.columns
```

```python
import pandas as pd

# Assuming spotify_df is already loaded
# Load your filenames dataset if needed
# filenames_df = pd.read_csv('filenames.csv')

# Remove any quotes from the column names if present


# Step 1: Clean the data in the columns if they include quotes
spotify_df['artistname'] = spotify_df['artistname'].str.replace('"', '')
spotify_df['trackname'] = spotify_df['trackname'].str.replace('"', '')

# Step 2 and 3: Concatenate and replace spaces with underscores
spotify_df['liked_songs'] = (spotify_df['artistname'] + '_-_' + spotify_df['trackname']).str.replace(' ', '_')

# Now your spotify_df has a new column 'formatted_name' that should match the format in filenames.csv

```

```python
spotify_df.to_csv("processed_user_spotify_dataset.csv",index=False)
```

```python
# If not already loaded
filenames_df = pd.read_csv('filenames.csv', header=None)  # Assuming there's no header
filenames_df.columns = ['song']  # Name the column for easier reference
```

```python
filename_set = set(filenames_df['song'])
```

```python
spotify_df = spotify_df[spotify_df['liked_songs'].isin(filename_set)]
```

```python
spotify_df
```

```python
# Factorize the user_id to get a unique integer for each unique hash ID
spotify_df['user_id'], _ = pd.factorize(spotify_df['user_id'])

# Convert these integers to 'user1', 'user2', etc.
spotify_df['user_id'] = 'user' + (spotify_df['user_id'] + 1).astype(str)
```

```python
spotify_df.to_csv("processed_user_spotify_dataset.csv",index=False)
```
