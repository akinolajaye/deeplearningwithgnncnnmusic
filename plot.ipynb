```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load data from a text file
# Make sure the file 'training_data.txt' is in your current working directory or provide the full path.
file_path = 'GNN_LOSS.txt'
df = pd.read_csv(file_path,header=None)
```

```python
# Define a function to clean and split the data entries
def clean_split(row):
    # Split each entry by colon and take the second part (the numeric value)
    epoch = int(row[0].split(':')[1].strip())
    train_loss = float(row[1].split(':')[1].strip())
    val_auc = float(row[2].split(':')[1].strip())
    return epoch, train_loss, val_auc

# Apply the function to each row
df = df.apply(lambda row: clean_split(row), axis=1, result_type='expand')

# Rename columns for clarity
df.columns = ['Epoch', 'Train Loss', 'Val AUC']

# Now you can plot or analyze your cleaned data
print(df)
```

```python

# Convert the 'Epoch' column to integers if it's not already
df['Epoch'] = pd.to_numeric(df['Epoch'], errors='coerce')

# Step 2: Plotting
plt.figure(figsize=(14, 6))

# Plot for Training Loss
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(df['Epoch'], df['Train Loss'], marker='o', linestyle='-', color='b',markersize=0.1)
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.grid(True)

# Plot for Validation AUC
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(df['Epoch'], df['Val AUC'], marker='o', linestyle='-', color='g',markersize=0.5)
plt.title('Validation AUC over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Validation AUC')
plt.grid(True)

plt.tight_layout()
plt.show()
```
